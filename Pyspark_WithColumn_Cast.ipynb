```python
from pyspark.sql import SparkSession
import pyspark.sql.functions as F
```

```python
spark = (
    SparkSession.builder
      .master('local')
      .appName('dataframe_withcolumn')
      .config('spark.ui.port', '4050')
      .getOrCreate()
)
```

```python
df1 = (spark
       .read
       .format("csv")
       .option("header", "true")
       .option("inferschema", "true")
       .option("delimiter", ";")
       .load('/content/drive/MyDrive/Colab Notebooks/arquivo_geral.csv')
)
```

```python
df1.printSchema()
```

```python
df1.show(20)
```

CRIAR UMA NOVA COLUNA COM UM VALOR ALEATÓRIO


```python
numero = 10
numero += 5
numero2 = numero + 5

valor = 25
df1 = df1.withColumn("novo_valor", F.lit(valor))
df1.show(50)
```

```python
df1.printSchema()
```

```python
df_novo_cast = df1.select(F.col("novo_valor").cast("string"))

df_novo_cast.printSchema()
```

```python
valor = "novo!"
df1 = df1.withColumn("NOVO MESMO?", F.lit(valor))
df1.show(50)
```
