```python
from pyspark.sql import SparkSession
import pyspark.sql.functions as F
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, DoubleType
```

```python
spark = (
    SparkSession.builder
      .master('local')
      .appName('dataframe_withcolumn')
      .config('spark.ui.port', '4050')
      .getOrCreate()
)
```

```python
df1 = (spark
       .read
       .format("csv")
       .option("header", "false")
       .option("inferschema", "false")
       .option("delimiter", ";")
       .load('/content/drive/MyDrive/Colab Notebooks/arquivo_geral.csv')
)
```

TERCEIRA MANEIRA - UTILIZANDO O SELECT COM F.COL E ALIAS


```python
df2 = (spark
       .read
       .format("csv")
       .option("header", "false")
       .option("inferschema", "false")
       .option("delimiter", ";")
       .load('/content/drive/MyDrive/Colab Notebooks/arquivo_geral.csv')
)
```

```python
df2 = df2.select(F.col("_c0").alias("um"),
                 F.col("_c1").alias("dois"),
                 F.col("_c2").alias("tres"),
                 F.col("_c3").alias("quatro"),
                 F.col("_c4").alias("cinco"),
                 F.col("_c5").alias("seis"),
                 F.col("_c6").alias("sete"),
)

df2.printSchema()
```

```python
df2.show(20)
```
